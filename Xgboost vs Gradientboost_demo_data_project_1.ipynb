```python
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
import time
```

```python
x, y = make_classification(n_samples=1000, n_features=200, random_state=42)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
gb_model = GradientBoostingClassifier( random_state=42)
xgb_model = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
```

```python
start_gb = time.time()
gb_model.fit(x_train, y_train)
end_gb = time.time()
gb_time = end_gb - start_gb
```

```python
gb_pred = gb_model.predict(x_test)
gb_accuracy = accuracy_score(y_test, gb_pred)
```

```python
start_xgb = time.time()
xgb_model.fit(x_train, y_train)
end_xgb = time.time()
xgb_time = end_xgb - start_xgb
```

```python
xgb_pred = xgb_model.predict(x_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)
```

```python
gb_time , gb_accuracy, xgb_time , xgb_accuracy
# as we clearly see that time diffrence gb: 7.9 , xgb: 1.9
# and accuracy in also littel incease   gb: 84% , xgb: 87%
```

```python
# let see
import numpy as np
import pandas as pd

X = np.array([[1,2],
              [3, np.nan],   # there are two value and i will not change it
              [7,8],
              [np.nan,6],
              [3,4],
              [8,9]])
Y = np.array([0,1,1,0,1,0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model_1 = xgb.XGBClassifier( use_label_encoder=False, eval_metric='logloss')
model_1.fit(X_train, y_train)
```

```python
y_pred = model_1.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy   # can you see that i don't do any preprossesing!!!!!!!!!!!
           # still it gie me rerult
```

```python
model_2 = GradientBoostingClassifier()
```

```python
model_2.fit(X_train, y_train)  # we can see that it rase an error for missing value
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer to replace NaN values with the mean
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform it
X_train_imputed = imputer.fit_transform(X_train)

# Now, fit the model on the imputed data
model_2.fit(X_train_imputed, y_train)

# To make predictions on the test set, you'll need to transform it as well
X_test_imputed = imputer.transform(X_test)
y_pred_2 = model_2.predict(X_test_imputed)

# Now you can calculate the accuracy
accuracy_2 = accuracy_score(y_test, y_pred_2)
print(f"Accuracy of GradientBoostingClassifier after imputation: {accuracy_2}")
```
